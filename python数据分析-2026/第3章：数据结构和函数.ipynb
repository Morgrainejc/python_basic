## 3.2 函数

```python
def my_function(x, y, z=1.5):
    if z > 1:
        return z * (x + y)
    else:
        return z / (x + y)
```

```python
my_function(1,3,5)
```

```python
my_function(2,4)
```

```python
def func():
    a = []
    for i in range(5):
        a.append(i)
    return a
```

#### 3.2.4 匿名函数

### 3.2.6 生成器

```python
def squares(n=10):
    print('Generating squares from 1 to {0}'.format(n ** 2))
    for i in range(1, n + 1):
        yield i ** 2
```

```python
gen = squares()
gen
```

```python
for x in gen:
    print(x, end=' ')
```

### 3.2.7 生成器表达式

```python
gen = (x ** 2 for x in range(100))
gen
```

```python
def _make_gen():
    for x in range(10):
        yield x ** 2
gen = _make_gen()
```

```python
for x in gen:
    print(x)
```
