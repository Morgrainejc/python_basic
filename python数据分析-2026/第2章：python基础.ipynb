```python
template = '{0:.2f} {1:s} are worth US${2:d}'
```

```python
template.format(4.5560, 'Argentine Pesos', 1)
```

```python
name = "苹果"
weight = 3
price = 5
total = weight * price  # 16.5

# 写法一：老式 % （占位符用 %s, %d, %.2f）
print("1. 老式%% ：%s %d斤，总价 %.2f元" % (name, weight, total))

# 写法二：.format() （就是你给的那种模板）
print("2. .format：{0:s} {1:d}斤，总价 {2:.2f}元".format(name, weight, total))

# 写法三：f-string （最直观，变量直接放进去）
print(f"3. f-string：{name} {weight}斤，总价 {total:.2f}元")
```
